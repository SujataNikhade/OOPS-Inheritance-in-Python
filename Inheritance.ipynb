## multiple inheritance

```python
#Create two classes Father and Mother each with unique methods.
class Father:
    def __init__(self,f_name,f_age,occupation):
        self.f_name=f_name
        self.f_age=f_age
        self.occupation=occupation
    def work(self):
        print("father name: ",self.f_name)
        print("age of my father:",self.f_age)
        print("occupation of my father: ",self.occupation)
class Mother:
    def __init__(self,m_name,m_age,hobby):
        self.m_name=m_name
        self.m_age=m_age
        self.hobby=hobby
    def care(self):
        print("Monther name:",self.m_name)
        print("mother age:",self.m_age)
        print("hobbies:",self.hobby)
class child(Father,Mother):
    def __init__(self,f_name,f_age,occupation,m_name,m_age,hobby,school,grade):
        Father.__init__(self,f_name,f_age,occupation)
        Mother.__init__(self,m_name,m_age,hobby)
        self.school=school
        self.grade=grade
    def display(self):
        self.work()
        self.care()
        print("school name: ",self.school)
        print("grade:",self.grade)

s=child("suresh",49,"employee","suchita",44,"cooking specealist","LLHSC","A+")
s.display()
```

```python
#Make classes Teacher (method teach()) and Coach (method train()).

class Teacher:
    def __init__(self,salary,experience):
        self.salary=salary
        self.experience=experience
    def teach(self):
        print("teacher salary: ",self.salary)
        print("teach id: ",self.experience)
class coach:
    def __init__(self,name,sport):
        self.sport=sport
        self.name=name
    def train(self):
        print("The team of coach: ",self.name)
        print("The sport:",self.sport)
class Mentor(Teacher,coach):
    def __init__(self,salary,experience,name,sport,level):
        Teacher.__init__(self,salary,experience)
        coach.__init__(self,name,sport)
        self.level=level
    def display(self):
        self.teach()
        self.train()
        print("level:",self.level)

s=Mentor(3000,"3-year","mayur","cricket","senior")
s.display()
```

```python
#Build Artist (method paint()) and Musician (method play_music()).
class Artist:
    def __init__(self,name,style,experience):
        self.name=name
        self.style=style
        self.experience=experience
    def paint(self):
        print("Artist name: ",self.name)
        print("art style:",self.style)
        print("experience on art:",self.experience)
class Musician:
    def __init__(self,instrument,genre):
        self.instrument=instrument
        self.genre=genre
    def play_music(self):
        print("Instrument are:",self.instrument)
        print("Genre are: ",self.genre)
class performer(Artist,Musician):
    def __init__(self,name,style,experience,instrument,genre,event):
        Artist.__init__(self,name,style,experience)
        Musician.__init__(self,instrument,genre)
        self.event=event
    def display(self):
        self.paint()
        self.play_music()
        print("which event:",self.event)

s=performer("sujju","Abstract","5year","Guitar","pop","art & music fest")
s.display()
```

```python
# Design FrontEnd and BackEnd classes. Create FullStackDeveloper that inherits both and prints all skills.
class Frout_end:
    def __init__(self,frontend_language,framework):
        self.frontend_language=frontend_language
        self.framework=framework
class Backend:
    def __init__(self,backend_lnguage,database):
        self.backend_lnguage=backend_lnguage
        self.database=database
class Fullstackdeveloper(Frout_end,Backend):
    def __init__(self,frontend_language,framework,backend_lnguage,database,name,exp):
        Frout_end.__init__(self,frontend_language,framework)
        Backend.__init__(self,backend_lnguage,database)
        self.name=name
        self.exp=exp
    def display(self):
        print("frontend language: ",self.frontend_language)
        print("framework is:",self.framework)
        print("backend language: ",self.backend_lnguage)
        print("data base:",self.database)
        print("Name:",self.name)
        print("experince:",self.exp)
    
s=Fullstackdeveloper("javscrip","React","python","Mysql","sujju","3-year")
s.display()
```

```python
## class Employee with attributes like name.Then subclasses: Engineer, Manager, and Intern. Each has a unique method.
class Employee:
    def __init__(self,name,emply_id):
        self.name=name
        self.emply_id=emply_id
    def emply_info(self):
        print("emplyee id:",self.emply_id)
        print("emplyee name :",self.name)
        
class Engineer(Employee):
    def __init__(self,name,emply_id,salary,project_name):
        super().__init__(name,emply_id)
        self.salary=salary
        self.project_name=project_name
    def eng_info(self):
        self.emply_info()
        print("engineer salary:", self.salary)
        print("project name:",self.project_name)
        
class Manager(Employee):
    def __init__(self,name,salary,emply_id,project_name,department,team):
        super().__init__(name,emply_id)
        self.department=department
        self.team=team
    def manager_info(self):
        self.emply_info()
        print("manger department:",self.department)
        print("Team member:",self.team)
        
class Internship(Employee):
    def __init__(self,name,salary,emply_id,project_name,department,duration,stipend):
        super().__init__(name,emply_id)
        self.duration=duration
        self.stipend=stipend
    def intern(self):
        self.emply_info()
        print("Duration in months:",self.duration)
        print("stipend:",self.stipend)
        
s=Internship("sujju",100000,"AB2010","AI automation","Data science","6-months",50000)
s.intern()
```

```python
## Define a class Appliance → children: Fan, WashingMachine, and Oven.
class Appliance:
    def __init__(self,name,brand,power_rating):
        self.name=name
        self.brand=brand
        self.power_rating=power_rating
    def power_usage(self):
        print("Name:",self.name)
        print("Brand:",self.brand)
        print("power rating:",self.power_rating)
        
class Fan(Appliance):
    def __init__(self,name,brand,power_rating,speed,blade_size,power_rate):
        super().__init__(name,brand,power_rating)
        self.speed=speed
        self.blade_size=blade_size
        self.power_rate=power_rate
        
    def power_usage(self):
        self.power_usage()
        print("speed:",self.speed)
        print("balde size:",self.blade_size)
        print("power rating:",self.power_rate)

class Washing_machine(Appliance):
    def __init__(self,name,brand,power_rating,load_capacity,wash_modes):
        super().__init__(name,brand,power_rating)
        self.load_capacity=load_capacity
        self.wash_modes=wash_modes
    def power_usage(self):
        self.power_usage()
        print("load capacity:",self.load_capacity)
        print("washing mode:",self.wash_modes)
class Oven(Appliance):
    def __init__(self,name,brand,power_rating,temp_range,capacity,power_use):
        super().__init__(name,brand,power_rating)
        self.temp_range=temp_range
        self.capacity=capacity
        self.power_use=power_use
    def power_usage(self):
        super().power_usage()
        print("temperature range:",self.temp_range)
        print("capacity:",self.capacity)
        print("power usage:",self.power_use)

c=Oven("Home Appliance", "Samsung", "1500W", "100°C–250°C", "30 Liters", "1000W per cycle")
c.power_usage()
```

```python
## Create a class Shape with a draw() method.
class Shape:
    def __init__(self,name):
        self.name=name
    def draw(self):
        print(f"Drawing a {self.name}")

class Circle(Shape):
    def __init__(self,name,radius):
        super().__init__(name)
        self.radius=radius
    def draw(self):
        super().draw()
        print(f"Area :{3.14*self.radius**2}")
        print(f"perimeter:{2*3.14*self.radius}")
class square(Shape):
    def __init__(self,name,length):
        super().__init__(name)
        self.length=length
    def draw(self):
        super().draw()
        print("Area", self.length**2)
        print("Perimeter:", self.length*4)
        
class Triangle(Shape):
    def __init__(self,name,base,height,sides):
        super().__init__(name)
        self.base=base
        self.height=height
        self.sides=sides
    def draw(self):
        super().draw()
        print(f"Area:",0.5*self.base*self.height)
        print(f"Paramenter:",sum(self.sides))
    
s=Triangle("Triangle",6,8,(2,3,4))
s.draw()

t=Circle("Circle",7)
t.draw()

c=square("square",6)
c.draw()
```

## hybrid inheritance

```python
## Create classes Person, Teacher, Researcher, and Professor.
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
class Teacher(Person):
    def __init__(self,name,age,salary):
        Person.__init__(self,name,age)
        self.salary=salary
        
class Researcher(Person):
    def __init__(self,name,age,salary,project):
        Person.__init__(self,name,age)
        self.salary=salary
        self.project=project
    def display(self):
        print("name:",self.name)
        print("age:",self.age)
        print("salary:",self.salary)
        print("project name:",self.project)
        
class Professor(Teacher,Researcher):
    def __init__(self,name,age,salary,project,department):
        Teacher.__init__(self,name,age,salary)
        Researcher.__init__(self,name,age,salary,project)
        self.department=department
    def display1(self):
        self.display()
        print("department:",self.department)
c=Professor("suuju",21,100000,"ai base","data science")
c.display1()
```

```python
# Design a structure: Vehicle → Car, Boat → AmphibiousVehicle
class vehicle:
    def __init__(self,name):
        self.name=name
class car(vehicle):
    def drive(self):
        print(self.name," id driving on the road")
class Boat(vehicle):
    def sail(self):
        print(self.name," is sailing on the water")
class Amphibiousvehicle(car,Boat):
    def __init__(self,name):
        super().__init__(name)
        
s=Amphibiousvehicle("AmphiX")
s.drive()
Amphibiousvehicle("sujju").sail()
```

```python
## Build classes Device → Computer and Phone → Smartphone
class Derice:
    def __init__(self,brand):
        self.brand=brand
class computer(Derice):
    def compute(self):
        print(self.brand," computer is processing the data")
class phone(Derice):
    def call(self):
        print(self.brand," phone is making a cell")
class smartphone(computer,phone):
    def __init__(self,brand):
        super().__init__(brand)
    def full_functionality(self):
        self.compute()
        self.call()
        print(self.brand, "smartphone combines computer and phone features")
    
s=smartphone("Tech")
s.full_functionality()
```
